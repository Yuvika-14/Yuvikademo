```python
df=df[['id','original_title','overview','genres']]
```

```python
df['tags']=df['overview']+df['genres']
```

```python
df['tags'] = df['overview']+df['genres']
```

```python
df
```

```python
new_data  =df.drop(columns=['overview', 'genres'])
```

```python
new_data
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=10000, stop_words='english')
```

```python
cv
```

```python

vector=cv.fit_transform(new_data['tags'].values.astype('U')).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
similarity
```

```python
new_data[new_data['original_title']=="Avatar"].index[0]
```

```python
distance = sorted(list(enumerate(similarity[2])), reverse=True, key=lambda vector:vector[1])
for i in distance[0:5]:
    print(new_data.iloc[i[0]].original_title)
```

```python
def recommand(movies):
    index=new_data[new_data['original_title']==movies].index[0]
    distance = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda vector:vector[1])
    for i in distance[0:5]:
        print(new_data.iloc[i[0]].original_title)
```

```python
recommand("Iron Man")
```
